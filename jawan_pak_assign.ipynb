```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon[1].csv")
data.head(n=10) #by default the head function shows only 5 rows
```

```python
speed_mean = data["Speed"].mean()

def set_speed(value):
    if value < speed_mean:
        return "Low Speed"
    elif value == speed_mean:
        return "Neutral Speed"
    else:
        return "High Speed"
```

```python
hp_mean = data["HP"].mean()

def set_HP(hp):
  if hp < hp_mean:
    return "Low HP"
  elif hp==hp_mean:
    return "Neutral HP"
  else:
    return "High HP"
```

```python
# "Speed_high_low"  -> "High Speed" , "Low Speed"
data["High_Low_Speed"] = data["Speed"].apply(set_speed)
data
```

```python
# "Speed_high_low"  -> "High Speed" , "Low Speed"
data["High_Low_HP"] = data["HP"].apply(set_HP)
data
```
